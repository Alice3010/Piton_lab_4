```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from sklearn.model_selection import cross_val_score, GridSearchCV, train_test_split 
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("full.csv")
```

```python
lb = LabelEncoder()
data['Embarked'] = lb.fit_transform(data['Embarked'].astype(str))
data['Sex'] = lb.fit_transform(data['Sex'].astype(str))
data['Cabin'] = lb.fit_transform(data['Cabin'].astype(str))
data = data.fillna(data.median())
```

```python
#full_data = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin']]
full_data = data[['Sex', 'Age']]
full_sur = data[['Survived']]
type(full_sur)
```

```python
train, test, train_sur, test_sur = train_test_split(full_data, full_sur, train_size=0.70, test_size=0.30)
```

```python
##Gaussian
#clf = GaussianNB()
#clf.fit(train, train_sur)
#print("NB Accuracy: "+repr(round(clf.score(test, test_sur) * 100, 2)) + "%")
#result_rf=cross_val_score(clf,train,train_sur,cv=10)
#print('The cross validated score for GNB is:',round(result_rf.mean()*100,2))
```

```python
##Gaussian
clf = GaussianNB()
params = { 'priors': [[0.4, 0.6],[0.5,0.5]], 
          'var_smoothing':[1e-12, 1e-15, 1e-9, 1e-5, 1e-4]}
gb_grid = GridSearchCV(clf, params, cv=30, refit = True)
```

```python
gb_grid.fit(train, train_sur)
print(gb_grid.best_params_)
print(gb_grid.best_score_)
print(gb_grid.score(test, test_sur))
```

```python
result = gb_grid.cv_results_
mass = result['params']
print(result)
mass0 = result['mean_test_score']
dataGB = pd.DataFrame(mass0, mass)
dataGB.plot(kind = 'line', title = 'score', figsize = (25,5), legend = False, rot = 90)
```

```python
##lLogisticRegression
clf = LogisticRegression()
clf.fit(train, train_sur)
print("NB Accuracy: "+repr(round(clf.score(test, test_sur) * 100, 2)) + "%")
result_rf=cross_val_score(clf,train,train_sur,cv=10)
print('The cross validated score for GNB is:',round(result_rf.mean()*100,2))
```

```python
clf = LogisticRegression()
lr_params = { 'penalty':('l1', 'l2'), 'C': range(1,11)}
lr_grid = GridSearchCV(clf, lr_params, cv=30, refit = True)
```

```python
lr_grid.fit(train, train_sur)
print(lr_grid.best_params_)
print(lr_grid.best_score_)
print(lr_grid.score(test, test_sur))
```
